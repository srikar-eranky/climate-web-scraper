```python
def change_link(link):
    new_link = link[:link.index("view")] + "downloadpdf/" + link[link.index("view"):link.index(".xml")] + ".pdf"
    return new_link
```

```python
print(change_link("https://journals.ametsoc.org/view/journals/clim/33/24/jcliD200801.xml"))
```

```python
df['Download Link'] = df['PDF Link'].apply(change_link)
df
```
